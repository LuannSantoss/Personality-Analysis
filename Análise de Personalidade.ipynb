# Creating clusters 

```python
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(dataset)
```

```python
predicoes = k_fit.labels_
dataset['Clusters'] = predicoes
```

```python
dataset.head()
```

```python
dataset['Clusters'].value_counts()
```

```python
dataset.groupby('Clusters').mean()
```

```python
col_list = list(dataset)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]
```

```python
data_soma = pd.DataFrame()
data_soma['extroversion'] = dataset[ext].sum(axis=1)/10
data_soma['neurotic'] = dataset[est].sum(axis=1)/10
data_soma['agreeable'] = dataset[agr].sum(axis=1)/10
data_soma['conscientious'] = dataset[csn].sum(axis=1)/10
data_soma['open'] = dataset[opn].sum(axis=1)/10
data_soma['clusters'] = predicoes
```

```python
data_soma.groupby('clusters').mean()
```

```python
data_clusters = data_soma.groupby('clusters').mean()
```

```python
plt.figure(figsize=(15,5))
for i in range(0,5):
    plt.subplot(1,5,i+1)
    plt.bar(data_clusters.columns, data_clusters.iloc[:,i], color='green', alpha=0.2)
    plt.plot(data_clusters.columns, data_clusters.iloc[:,i], color='red')
    plt.title('Grupo '+str(i))
    plt.xticks(rotation = 45)
    plt.ylim(0,4);
```

# Deploy

```python
import gradio as gr
```

```python
dicio_questions = open('C:/Users/luann/Arquivos Datasets/questions.txt').read().split('\n')
```

```python
dicio_questions
```

```python
questions = []
for q in dicio_questions:
    q = str(q)
    questions.append(q[q.find('\t'):].lstrip())
```

```python
questions
```

```python
inputs_questions = []
for q in questions:
    obj_input = gr.inputs.Slider(minimum=1, maximum=5, step=1, default=3, label=q)
    inputs_questions.append(obj_input)
```

```python
inputs_questions
```

```python
def predict(*outputs_questions):
    outputs_questions = np.array(outputs_questions).reshape(1,-1)
    return k_fit.predict(outputs_questions)
iface = gr.Interface(
                    fn = predict,
                    title = 'Big Five Personality',
                    description = 'Sistema para deteco de traos de personalidade',
                    inputs = inputs_questions,
                    outputs = 'text')
iface.launch(share=False)
```
